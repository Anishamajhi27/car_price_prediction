```python
df = pd.read_csv("CarPrice_Assignment.csv")
df.head()
```

```python
category = df.dtypes == "object"
category = df.columns[category]
category
```

```python
for i in range (0,len(category)):
  df[category[i]] = df[category[i]].astype('category')
  df[category[i]] = df[category[i]].cat.codes
df.head()
```

```python
df.dtypes == "object"
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns = "price")
x.head()
```

```python
y = df["price"]
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
y_pred_train = lr.predict(X_train)
y_pred_train
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train,y_pred_train)
plt.xlabel("Y_train")
plt.ylabel("Y_pred_train")
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,y_pred_train)*100
```

```python
y_pred_test = lr.predict(X_test)
y_pred_test
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred_test)
plt.xlabel("Y_test")
plt.ylabel("Y_pred_test")
```

```python
r2_score(y_test,y_pred_test)*100
```
